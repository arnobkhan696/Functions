1. What is the difference between a function and a method in Python ?
  - In Python, a function is a reusable block of code that performs a specific task.
  - While a method is a function associated with an object or class, allowing it to operate on the object's data

```python
#Example: Function
def greet(name):
  return f"Hello, {name}!"
print(greet("Arnob"))

```

```python
# Method
class Greeter:
    def greet(self, name):
        return f"Hello, {name}!"

greeter = Greeter()
print(greeter.greet("Arnob"))
```

6. Explain the concept of generators in Python and how they are defined ?
  - A generator is a special type of iterator in Python that is used to generate values on the fly without storing them all in memory at once.
Generators are memory-efficient and are used when you want to iterate over a large set of data but don’t want to store it all at once

```python
# Example:
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

# Using the generator
gen = count_up_to(3)
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
print(next(gen))  # Output: 3
```

7. What are the advantages of using generators over regular functions ?
  - Unlike a regular function, a generator does not return its results all at once. Instead, it yields its values one by one, each time it is called. This makes it possible to generate an infinite sequence of values, as long as there is sufficient memory to store them.

10. What is the difference between "map()", "reduce()", and "filter()" functions in Python?
   - The map() function is used to apply a specific function to every element in an iterable (like a list or tuple).
It transforms each element and returns a map object (an iterator) with the modified elements.
   - The filter() function is used to filter elements of an iterable based on a condition.
It returns a filter object (an iterator) containing only those elements that satisfy the condition.
   - The reduce() function is used to apply a function cumulatively to all elements of an iterable, reducing them to a single value.

```python
# Example:
from functools import reduce

numbers = [1, 2, 3, 4, 5]

# map() → Double each number
result_map = list(map(lambda x: x * 2, numbers))
print(result_map)  # Output: [2, 4, 6, 8, 10]

# filter() → Keep even numbers
result_filter = list(filter(lambda x: x % 2 == 0, numbers))
print(result_filter)  # Output: [2, 4]

# reduce() → Sum of all numbers
result_reduce = reduce(lambda x, y: x + y, numbers)
print(result_reduce)  # Output: 15
```

```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_of_even(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum

my_list = [10, 3, 4, 7, 8, 9, 12]
result = sum_of_even(my_list)
print("Sum of even numbers:", result)
```

```python
# 2. Create a Python function that accepts a string and returns the reverse of that string.
def reverse_string(text):
    return text[::-1]

input_string = "Hello World"
reversed_string = reverse_string(input_string)
print("Reversed string:", reversed_string)
```

```python
# 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
def square_numbers(numbers):
    squared_list = []
    for num in numbers:
        squared_list.append(num ** 2)
    return squared_list

# Example usage
my_list = [2, 3, 4, 5]
result = square_numbers(my_list)
print("Squared numbers:", result)
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

for num in range(1, 201):
    if is_prime(num):
        print(num, end=" ")
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class Fibonacci:
    def __init__(self, terms):
        self.terms = terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

fib = Fibonacci(10)
for num in fib:
    print(num, end=" ")
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

for num in power_of_two(5):
    print(num, end=" ")
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_lines(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()

for line in read_file_lines('sample.txt'):
    print(line)
```

```python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
data = [(1, 5), (3, 2), (4, 8), (2, 1)]

sorted_data = sorted(data, key=lambda x: x[1])

print("Sorted list:", sorted_data)
```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
celsius = [0, 20, 37, 100]

fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print("Temperatures in Fahrenheit:", fahrenheit)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def is_not_vowel(char):
    return char.lower() not in 'aeiou'

text = "Hello World, Python is Fun!"

result = ''.join(filter(is_not_vowel, text))

print("String without vowels:", result)
```
